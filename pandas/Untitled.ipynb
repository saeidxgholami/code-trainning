```python
import pandas as pd
import calc


df = pd.read_csv('biostats.csv', skipinitialspace=True)
# 'Height (in)', 'Weight (lbs)'

heights = df['Height (in)']
weights = df['Weight (lbs)']

height_m = []
weight_kg = []
bmi = []
bmi_status = []

for height, weight in zip(heights, weights):
    height_m.append(calc.meter(height))
    weight_kg.append(calc.kg(weight))

# claculate bmi
for h, w in zip(height_m, weight_kg):
    bmi.append(calc.bmi(h, w))

for b in bmi:
    bmi_status.append(calc.bmi_status(b))

result = {
    'height': [],
    'weight': [],
    'bmi': [],
    'bmi_status': []
}
for h, w, b, bs in zip(height_m, weight_kg, bmi, bmi_status):
    result['height'].append(h)
    result['weight'].append(w)    
    result['bmi'].append(b)    
    result['bmi_status'].append(bs)    

result_df = pd.DataFrame(result)
result_df.to_csv('bmi.csv', index=False)
x = pd.read_csv('bmi.csv')
x

```
